## 0. Setup & Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter

import ssl
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context
import nltk
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('All imports OK!')
```

## 1. Data Collection

```python
df_raw = pd.read_csv('mbti_1.csv')

print(f'Shape: {df_raw.shape}')
print(f'Columns: {list(df_raw.columns)}')
df_raw.head()
```

```python
# Basic info
print(df_raw.info())
print(f'\nNull values:\n{df_raw.isnull().sum()}')
print(f'\nUnique MBTI types: {df_raw["type"].nunique()}')
print(df_raw['type'].unique())
```

## 2. Data Cleaning

### 2.1 Split posts (each user has ~50 posts separated by `|||`)

```python
df_raw['posts_list'] = df_raw['posts'].str.split('\\|\\|\\|')
df_raw['num_posts'] = df_raw['posts_list'].apply(len)

print(f'Posts per user - min: {df_raw["num_posts"].min()}, '
      f'max: {df_raw["num_posts"].max()}, '
      f'mean: {df_raw["num_posts"].mean():.1f}')
```

### 2.2 Text preprocessing pipeline

```python
punctuations = ''''"\\,<>./?@#$%^&*_~/!()-[]{};:'''
stop_words = set(stopwords.words('english'))
wln = WordNetLemmatizer()

mbti_types = [
    'infp', 'infj', 'intp', 'intj', 'entp', 'enfp', 'istp', 'isfp',
    'entj', 'istj', 'enfj', 'isfj', 'estp', 'esfp', 'esfj', 'estj'
]
stop_words.update(mbti_types)

def clean_text(text):
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = text.replace('|||', ' ')
    for p in punctuations:
        text = text.replace(p, ' ')
    text = text.strip()
    text = text.lower()
    words = text.split()
    words = [wln.lemmatize(w) for w in words if w not in stop_words and len(w) > 1]
    return ' '.join(words)

print('Cleaning text... (takes ~1 min)')
df_raw['clean_posts'] = df_raw['posts'].apply(clean_text)
print('Done!')

# Sanity check
print('\n--- Raw (first 300 chars) ---')
print(df_raw['posts'].iloc[0][:300])
print('\n--- Cleaned (first 300 chars) ---')
print(df_raw['clean_posts'].iloc[0][:300])
```

```python
df_raw['word_count_raw'] = df_raw['posts'].str.split().apply(len)
df_raw['word_count_clean'] = df_raw['clean_posts'].str.split().apply(len)

df_raw['IE'] = df_raw['type'].apply(lambda x: x[0])
df_raw['SN'] = df_raw['type'].apply(lambda x: x[1])
df_raw['TF'] = df_raw['type'].apply(lambda x: x[2])
df_raw['JP'] = df_raw['type'].apply(lambda x: x[3])

df_raw.head()
```

## 3. Exploratory Data Analysis (EDA)

### 3.4 Top discriminative words per dichotomy

```python
def top_words_by_group(df, col, val1, val2, top_n=15):
    text1 = ' '.join(df[df[col] == val1]['clean_posts'].values).split()
    text2 = ' '.join(df[df[col] == val2]['clean_posts'].values).split()

    freq1 = Counter(text1)
    freq2 = Counter(text2)

    total1, total2 = sum(freq1.values()), sum(freq2.values())

    common = set(freq1.keys()) & set(freq2.keys())
    ratios = {}
    for w in common:
        r1 = freq1[w] / total1
        r2 = freq2[w] / total2
        if freq1[w] + freq2[w] > 50:
            ratios[w] = r1 / r2

    sorted_words = sorted(ratios.items(), key=lambda x: x[1], reverse=True)
    return sorted_words[:top_n], sorted_words[-top_n:]

for col, v1, v2, title in [
    ('IE', 'I', 'E', 'Introversion vs Extraversion'),
    ('SN', 'S', 'N', 'Sensing vs Intuition'),
    ('TF', 'T', 'F', 'Thinking vs Feeling'),
    ('JP', 'J', 'P', 'Judging vs Perceiving'),
]:
    top1, top2 = top_words_by_group(df_raw, col, v1, v2)
    print(f'\n=== {title} ===')
    print(f'  Top words for {v1}: {[w for w, r in top1]}')
    print(f'  Top words for {v2}: {[w for w, r in top2]}')
```

### 3.5 Average word count by dichotomy side

```python
summary = {}
for col in ['IE', 'SN', 'TF', 'JP']:
    summary[col] = df_raw.groupby(col)['word_count_clean'].mean()

summary_df = pd.DataFrame(summary).T
print('Average cleaned word count per dichotomy side:')
print(summary_df.round(1))
```

## 4. Train/Test Split (Stratified 80/20)

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(
    df_raw, test_size=0.2, random_state=42, stratify=df_raw['type']
)

print(f'Train: {len(df_train)}, Test: {len(df_test)}')
print(f'\nTrain type distribution (top 5):')
print(df_train['type'].value_counts().head())

cols_to_save = ['type', 'IE', 'SN', 'TF', 'JP', 'clean_posts', 'word_count_clean']
df_raw[cols_to_save].to_csv('mbti_cleaned.csv', index=False)
print(f'mbti_cleaned.csv: {len(df_raw)} rows')
df_train.to_csv('mbti_train.csv', index=False)
df_test.to_csv('mbti_test.csv', index=False)
print('Saved: mbti_train.csv, mbti_test.csv')
```

## 5. Summary of Findings

Fill in after running:

- **Dataset size:** 8675 users, ~50 posts each (min 1, max 89, mean 48.7)
- **Class imbalance:** INFP is the most common type (1832 users), ESTJ is the least common (39 users), imbalance ratio is 47.0x
- **Dichotomy imbalance:**  I (6676) >> E (1999), N (7478) >> S (1197), F (4694) > T (3981), P (5241) > J (3434). I/E and N/S are heavily skewed — typical of online forum data where introverts and intuitive types are overrepresented.
- **Post length patterns:** Average word count per user is roughly 500–700 words across all types, no major differences between types. Per-dichotomy differences are also small (I (594.5) vs E (593.7), N (598.0) vs S (571.4), F (603.1) vs T (584.0), J (599.4) vs P (591.0))
- **Discriminative words:** The T/F dimension shows the clearest lexical distinction T types tend to use analytical/technical vocabulary (e.g., inefficient, factual, calculation) while F types use more emotional/relational languag (e.g., dreamer, healing, idealize, empath). S/N also shows some separation — N types use more abstract words (glimpse, striving, enlighten) while S types use more concrete words (rave, streaming, eagle). I/E and J/P dimensions show weaker lexical signals.
- **Key challenge:**  Heavy class imbalance means the model will need stratified splits and macro F1 as the evaluation metric, not accuracy. MBTI type mentions in text must be removed to prevent data leakage.
